# 🔧 Notebook 4 — Feature Engineering
**Goal:** Create lag features, rolling averages, and interaction features to improve model accuracy.

---

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/cleaned_data.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(['Store ID','Product ID','Date'])
print('Shape:', df.shape)
df.head()
```

## Lag Features
**What:** Lag features tell the model what sales were N days ago. Very powerful for time series.

```python
# Sales from 1, 7, 14, 30 days ago
df['Sales_Lag1']  = df.groupby(['Store ID','Product ID'])['Units Sold'].shift(1)
df['Sales_Lag7']  = df.groupby(['Store ID','Product ID'])['Units Sold'].shift(7)
df['Sales_Lag14'] = df.groupby(['Store ID','Product ID'])['Units Sold'].shift(14)
df['Sales_Lag30'] = df.groupby(['Store ID','Product ID'])['Units Sold'].shift(30)
print('Lag features added!')
```

## Rolling Average Features
**What:** Rolling averages smooth out noise. They show the recent trend instead of a single noisy day.

```python
df['Rolling_7day_avg']  = df.groupby(['Store ID','Product ID'])['Units Sold'].transform(lambda x: x.shift(1).rolling(7).mean())
df['Rolling_14day_avg'] = df.groupby(['Store ID','Product ID'])['Units Sold'].transform(lambda x: x.shift(1).rolling(14).mean())
df['Rolling_30day_avg'] = df.groupby(['Store ID','Product ID'])['Units Sold'].transform(lambda x: x.shift(1).rolling(30).mean())
print('Rolling average features added!')
```

## Interaction Feature
**What:** Combine Store + Category as a single feature — each store may sell each category differently.

```python
from sklearn.preprocessing import LabelEncoder
df['Store_Cat'] = df['Store ID'].astype(str) + '_' + df['Category']
df['Store_Cat_enc'] = LabelEncoder().fit_transform(df['Store_Cat'])
print('Store-Category interaction feature added!')
```

## Drop NaN rows created by lag and save

```python
before = len(df)
df.dropna(inplace=True)
print(f'Rows dropped due to lag NaN: {before - len(df)}')
print(f'Final shape: {df.shape}')
df.to_csv('../data/featured_data.csv', index=False)
print('\n✅ Saved to ../data/featured_data.csv')
```
